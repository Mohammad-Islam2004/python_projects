```python
import pandas as pd
import numpy as np
import ast
```

```python
movies = pd.read_csv(r'C:\Users\jnj\Desktop\movie recommender\tmdb_5000_movies.csv')
credits = pd.read_csv(r'C:\Users\jnj\Desktop\movie recommender\tmdb_5000_credits.csv')
```

```python
movie = movies.merge(credits, on='title')
```

```python
movie.head(1)
```

```python
movie = movie[['movie_id','title','overview','genres','keywords','cast','crew']]
```

```python
movie.head(2)
```

```python
movie.isnull().sum()
```

```python
movie.dropna(inplace= True)
```

```python
movie.duplicated().sum()
```

```python
def transform(obj):
    L = []
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L
```

```python
movie['genres'] = movie['genres'].apply(transform)
movie["keywords"] = movie["keywords"].apply(transform)
```

```python
def transform1(obj):
    L = []
    counter = 0
    for i in ast.literal_eval(obj):
        if counter != 3:
            L.append(i["name"])
            counter += 1
        else :
            break
    return L
```

```python
movie['cast'] = movie['cast'].apply(transform1)
```

```python
def transform2(obj):
    L = []
    counter = 0
    for i in ast.literal_eval(obj):
        if i['job'] == 'Director':
            L.append(i["name"])
            break
    return L
```

```python
movie['crew'] = movie['crew'].apply(transform2)
```

```python
movie['overview'] = movie['overview'].apply(lambda x:x.split())
```

```python
movie.head()
```

```python
movie['genres'] = movie['genres'].apply(lambda x:[i.replace(" ","") for i in x])
movie['keywords'] = movie['keywords'].apply(lambda x:[i.replace(" ","") for i in x])
movie['cast'] = movie['cast'].apply(lambda x:[i.replace(" ","") for i in x])
movie['crew'] = movie['crew'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
movie.head()
```

```python
movie['tags'] = movie['overview'] + movie['genres'] + movie['keywords'] + movie['cast'] + movie['crew']
```

```python
movie.head()
```

```python
df = movie[['movie_id','title','tags']]
df
```

```python
df = df.copy()
df['tags'] = df['tags'].apply(
    lambda x: " ".join(x) if isinstance(x, list) else x
)
df
```

```python
df['tags'] = df['tags'].apply(lambda x:x.lower())
```

```python
import nltk
from nltk.stem.porter import PorterStemmer

ps = PorterStemmer()

def stem(text):
    y = []
    
    for i in text.split():
        y.append(ps.stem(i))
        
    return " ".join(y)

df['tags'] = df['tags'].apply(stem)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words='english')
vectors = cv.fit_transform(df['tags']).toarray()
```

```python
vectors
```

```python
cv.get_feature_names_out()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vectors)
```

```python
def recommend(movie):
    movie_index = df[df['title'] == movie].index[0]
    distances = similarity[movie_index]
    movies_list = sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[1:6]
    
    for i in movies_list:
        print(df.iloc[i[0]].title)
```

```python
recommend('Batman Begins')
```
